```python
#Importing the Relevant module
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#loading the data
df_data=pd.read_csv('homeprices.csv')
df_data
#Predicting Price of Plot
```

```python
df_data.bedrooms=df_data.bedrooms.fillna(df_data.bedrooms.median())
df_data
```

```python
new_data=df_data.drop('price',axis=1)
new_data
```

```python
df_price=df_data.price
df_price
```

```python
#Creating the regression model
reg = linear_model.LinearRegression()
reg.fit(new_data,df_price)
```

```python
#Predicting the price of 4000sq.ft with 2 bedroooms with age 40 yrs
reg.predict([[4000,2,40]])
```
